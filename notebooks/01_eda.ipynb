```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
print(df['Class'].value_counts())
```

```python
plt.figure(figsize=(14, 10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y=df['Amount'] + 0.01, data=df, palette='Set2')
plt.title('Transaction Amount Distribution by Class (Log Scale)')
plt.yscale('log') 
plt.ylabel('Amount in USD (log scale)')
plt.xlabel('Class (0: Normal, 1: Fraud)')
plt.show()
```

```python
print("--- Transaction Amount Statistics by Class ---")
pd.set_option('display.float_format', lambda x: '%.2f' % x)

display(df.groupby('Class')['Amount'].describe())
```
